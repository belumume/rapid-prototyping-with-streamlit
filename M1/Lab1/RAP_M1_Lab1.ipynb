```python
# update this with the location of the dataset on your computer
dataset_location = "/Users/summerrae/Rap/avalanche/data/customer_reviews.csv"

import pandas as pd

avalanche_df = pd.read_csv(dataset_location)
avalanche_df.head()
```

After reading in the dataset, we can see that it consists of customer reviews for a group of winter sport products. Each review containts a data, a summary of the review, and a sentiment score to determine if the review is positive or negative. 

### What products do we sell? 

```python
# what unique products do we sell? 
avalanche_df.PRODUCT.unique()
```

### What date range do we have data for? 

```python
# what date range are we working with?

# convert the date column to date_time format
avalanche_df["DATE"] = pd.to_datetime(avalanche_df["DATE"])

# Determine the date range
start_date = avalanche_df['DATE'].min()
end_date = avalanche_df['DATE'].max()

print(f"Start Date: {start_date}")
print(f"End Date: {end_date}")
```

### What is the range of sentiment scores? 

```python
# what is the range of sentiment scores? 

# Determine the date range
low_score = avalanche_df['SENTIMENT_SCORE'].min()
high_score = avalanche_df['SENTIMENT_SCORE'].max()

print(f"Low score: {low_score}")
print(f"High Score: {high_score}")
```
